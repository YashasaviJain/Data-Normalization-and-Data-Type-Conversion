Part A - Data Normilisation

```python
#Dataset
import pandas as pd
import numpy as np
data={'Product':['Laptop','Moblie','Tablet','Headphones','Smartwatch','Camera'],'Price':[55000,20000,30000,2000,8000,45000],'Units_Sold':[25,60,35,80,50,15,],'Discount':[10,15,12,20,18,8]}
df=pd.DataFrame(data)
print(df)
```

```python
#1. Min-Max Normalization
# Min-Max Normalization ffor Price
df['Price_Min_Max']=(df['Price']-df['Price'].min())/(df['Price'].max()-df['Price'].min())
print(df[['Product','Price','Price_Min_Max']])
print(df)
```

```python
#Normalize Multiple Columns
cols=['Price','Units_Sold','Discount']
df_norm=(df[cols]-df[cols].min())/(df[cols].max()-df[cols].min())
print(df)
print(df_norm)
```

```python
#2 Z-score normalization
df['Units_Z_Score']=(df['Units_Sold']-df['Units_Sold'].mean())/df['Units_Sold'].std()
print(df[['Product','Units_Sold','Units_Z_Score']])
```

```python
#Decimal Scaling
df['Price_Decimal_Scaled']=df['Price']/100000
print(df[['Product','Price','Price_Decimal_Scaled']])
```

Part B - Data Type Conversion

```python
#data encoding
import pandas as pd
import numpy as np
data = {
'Order_ID':[101,102,103,104,105,106],
'Customer_Gender':['Male','Female','Female','Male','Female','Male'],
'Payment_Method':['UPI','Credit Card','Debit Card','COD','UPI','Debit Card'],
'Product_Category':['Electronics','Clothing','Home','Electronics','Beauty','Electronics'],
'City':['Pune','Mumbai','Delhi','Bangalore','Pune','Hyderabad'],
'Order_Value':[4500,2300,3200,5200,1800,4100]
}

df1 = pd.DataFrame(data)

print(df1)
```

```python
#Method 1: Label Encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df1['Gender_Label']=le.fit_transform(df1['Customer_Gender'])
df1['City_Label']=le.fit_transform(df1['City'])
print(df1)
```

```python
#Method 2: One-Hot Encoding
df_encoded=pd.get_dummies(df1,columns=['Customer_Gender'])
print(df_encoded)
```

```python
#Method 4: Dummy Encoding
df1_encoded1 = pd.get_dummies(df1, columns = ['Payment_Method'],drop_first=True)
print(df1_encoded1)
```
